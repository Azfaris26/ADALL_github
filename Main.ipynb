```python
import pandas as pd
# Example: Replace this with the raw URL of your GitHub file
github_raw_url = 'https://raw.githubusercontent.com/Azfaris26/ADALL_github/refs/heads/main/laptop_prices_2024_sgd_TL.csv'
try:
  df = pd.read_csv(github_raw_url)
  print("Successfully loaded data from GitHub!")
  display(df.head())
except Exception as e:
  print(f"Error loading data: {e}")
  print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")
```

```python
print("Column names:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
# Exclude Price_SGD as it was already plotted
if 'Price_SGD' in numerical_cols:
    numerical_cols.remove('Price_SGD')

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

These histograms show the distributions of the various numerical features in the dataset. Examining these plots helps us understand the spread, central tendency, and any potential outliers or unusual patterns within each variable. This can inform subsequent data preprocessing and model building steps.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Price_SGD'], kde=True, bins=30)
plt.title('Distribution of Laptop Prices (SGD)')
plt.xlabel('Price (SGD)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

This histogram shows the distribution of laptop prices in SGD. We can observe the most common price ranges and identify if the distribution is skewed or has multiple peaks. We can then proceed to visualize relationships between `Price_SGD` and other features.
